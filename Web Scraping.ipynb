```python
url = 'https://www.worldometers.info/coronavirus/'
import requests
html = requests.get(url).content
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(html,'html.parser')
```

```python
soup
```

```python
tbody = soup.find('tbody')
```

```python
tbody
```

```python
table_rows = tbody.find_all('tr')
```

```python
table_rows
```

```python
table_rows[8]
```

```python
for i in range(8,len(table_rows)):
    td = table_rows[i].find_all('td')
    for table_data in td:
        print(table_data.text)
```

```python
all_data = []
for i in range(8,len(table_rows)):
    td = table_rows[i].find_all('td')
    
    country_data = []
    for table_data in td:
        country_data.append(table_data.text)
    all_data.append(country_data)
```

```python
all_data
```

```python
data = list(map(lambda x: x[1:7], all_data))
```

```python
data
```

```python
import pandas as pd
```

```python
pd.DataFrame(data)
```

```python
column = ['Country','Total Cases','New Cases', 'Total Death', ' New Death','Total Recovered']
```

```python
df = pd.DataFrame(data,columns = column)
```

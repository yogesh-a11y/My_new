```python
class Vehicle:
    def __init__(self, tyre,model):
        self.tyre = tyre
        self.model = model
        
    def show_tyre(self):
        return self.tyre
    
class Car(Vehicle):
    def __init__(self, color, weight, speed, name, tyre, model):
        self.color = color
        self.weight = weight
        self.speed = speed
        self.name = name
        super().__init__( tyre, model)
        
    def show_speed(self):
        return self.speed
    
    def show_height(self):
        return self.height
    
    
    
```

```python
obj = Car('red',7000, '150km/hr', 'Buggati', 4, 'Racing')
```

```python
obj.show_tyre() # calling with function 
```

```python
class Animal:
    def __init__(hello, legs, category): #we can replace self using any name
        hello.legs = legs
        hello.what = category # what le category ko value dinxa
        
    def show_category(hello):
        return hello.what
        
class Cow(Animal):
    def __init__(hello,color, tail, legs, category): 
        hello.color = color
        hello.tail = tail
        super().__init__(legs, category)
```

```python
color = 'Black'
tail = 1
legs = 4
category = 'Mammal'
```

```python
obj = Cow(color, tail, legs, category)
```
